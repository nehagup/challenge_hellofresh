```python
#importing libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
import statsmodels.tsa.stattools as ts
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
import seaborn as sns
```

```python
data = pd.read_csv(r"UKretail.csv", delimiter=",", encoding="latin1") #Reading data to Dataframe
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate']) #converting InvoiceDate column to datetime 
data_copy = data.copy()
data_copy['Revenue'] = data_copy.UnitPrice * data_copy.Quantity #Revenue column added
data_grouped = data_copy.groupby(pd.DatetimeIndex((data_copy['InvoiceDate'])).normalize())[['Revenue']].sum()
data_grouped.index.name = 'InvoiceDate'
data_grouped.to_csv("Revenue.csv" ,encoding= "utf-8", sep=",")
data_grouped.plot(figsize=(15,12))
plt.ylabel('Revenue')
```

```python
#ACF and PACF
diag = plt.figure(figsize=(15,12))
ax1 = diag.add_subplot(222)
diag = sm.graphics.tsa.plot_acf(data_grouped, lags=40, ax=ax1)
ax2 = diag.add_subplot(221)
diaf = sm.graphics.tsa.plot_pacf(data_grouped, lags=40, ax=ax2)
```
